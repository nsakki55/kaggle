```python
def categorical_col_labelencode(df_train, df_test):
    train_categorical_cols = df_train.select_dtypes(include='object').columns
    for col in train_categorical_cols:
        le=LabelEncoder()
        le.fit(list(df_train[col].astype(str).values)+list(df_test[col].astype(str).values))
        df_train[col]=le.transform(list(df_train[col].astype(str).values))
        df_test[col]=le.transform(list(df_test[col].astype(str).values))
    return df_train, df_test
```

```python
 def genres_parser(data):
        if len(data) == 2:
            return 'nan'
        else:
            return ','.join(x['name'] for x in ast.literal_eval(data))

def get_genres(df):
    df['genres'].fillna('[]', inplace=True)
    df['genre_name'] = df['genres'].apply(genres_parser)
    return df

train_merge = get_genres(train_merge)
```
