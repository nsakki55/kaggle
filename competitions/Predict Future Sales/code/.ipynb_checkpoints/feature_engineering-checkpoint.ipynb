```python
from tqdm import tqdm
for i in tqdm(np.unique(train["shop_id"])):
    for j in tqdm(np.unique(train['item_id'])):
        item_cnt_month=train.loc[train['date_block_num']==0][train['shop_id']==i][train['item_id']==j]['item_cnt_day'].sum()
        price=train.loc[train['date_block_num']==0][train['shop_id']==i][train['item_id']==j]["item_price"].values
        df_tmp=pd.DataFrame({'dat_block_num':[0],'shop_id':[i],'item_id':[j],"item_price":[price.mean()],"item_cnt_month":[item_cnt_month]})
        display(df_tmp)
        df=pd.concat([df,df_tmp],axis=0)
```
