```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display
import pandas_profiling as pdp
import japanize_matplotlib

import lightgbm as lgb
from sklearn.model_selection import train_test_split,KFold
from sklearn.metrics import accuracy_score
```

```python
X_train=pd.read_pickle('X_train_feature1.pkl')
X_test=pd.read_pickle('X_test_feature1.pkl')
Y_train=pd.read_pickle('y_train_feature1.pkl')
```

```python
x_train,x_valid,y_train,y_valid=train_test_split(X_train,Y_train,test_size=0.2,random_state=0)

lgb_train=lgb.Dataset(x_train,y_train)
lgb_valid=lgb.Dataset(x_valid,y_valid)

lgb_params={'objective':'binary'}

evals_result={}
model=lgb.train(params=lgb_params,
                             train_set=lgb_train,
                             valid_sets=[lgb_train,lgb_valid],
                             early_stopping_rounds=20,
                             evals_result=evals_result,
                             verbose_eval=10)

oof=(model.predict(x_valid)>0.5).astype(int)
print('score',round(accuracy_score(y_valid,oof)*100,2))
```

```python
plt.plot(evals_result['training']['binary_logloss'],label='train_loss')
plt.plot(evals_result['valid_1']['binary_logloss'],label='valid_loss')
plt.legend()
plt.show()
```

```python
feature_importance_df = pd.concat([
    pd.Series(X_train.columns),
    pd.Series(model.feature_importance())], axis=1)
feature_importance_df.columns = ['featureName', 'importance']

# get top 100 features sorted by importance descending
temp = feature_importance_df.sort_values(by=['importance'], ascending=False).head(100)

plt.figure(figsize=(15, 10))
sns.barplot(x="importance", y="featureName", data=temp)
plt.show()
```

k交差分割検証

```python
kfold=KFold(n_splits=3,shuffle=True)

score_list=[]
models=[]

for fold_,(train_index,valid_index) in enumerate(kfold.split(X_train,Y_train)):
    print(f'fold{fold_+1} start')
    x_train=X_train.iloc[train_index]
    x_val=X_train.iloc[valid_index]
    y_train=Y_train[train_index]
    y_valid=Y_train[valid_index]
    
    lgb_train=lgb.Dataset(x_train,y_train)
    lgb_valid=lgb.Dataset(x_valid,y_valid)
    
    lgb_params={'objective':'binary'}
    
    model=lgb.train(params=lgb_params,
                                 train_set=lgb_train,
                                 valid_sets=[lgb_train,lgb_valid],
                                 early_stopping_rounds=20,
                                 verbose_eval=-1)
    
    oof=(model.predict(lgb_valid)>0.5).astype(int)
    score_list.append(round(accuracy_score(y_valid,oof)*100,2))
    models.append(model)
    print(f'fold{fold_+1} end\n')

print('mean:',round(np.mean(score_list,2)))
```
