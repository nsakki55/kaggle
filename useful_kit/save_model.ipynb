# 学習したモデルを保存する
何時間もかけて学習したモデルをローカルに保存して、後から読み込める方法  
pythonのpickleを使う

```python
import pickle
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
data=load_boston()
X=data.data
y=data.target
cols=data.feature_names
df=pd.DataFrame(X,columns=cols)

X_train,X_test,y_train,y_test=train_test_split(df,y,test_size=.2,random_state=0)
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X_train_std=sc_x.fit_transform(X_train)
X_test_std=sc_x.transform(X_test)

y_train_std=sc_y.fit_transform(y_train[:,np.newaxis])
y_test_std=sc_y.fit_transform(y_test[:,np.newaxis])
```

## 学習モデルの作成

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
svm=SVR()
grid_dict=[{'kernel':['rbf'],'C':[10**i for i in range(-3,3)],'gamma':[10**i for i in range(-3,1)]}]
svm_reg=GridSearchCV(estimator=svm,param_grid=grid_dict,scoring='r2',cv=5)
svm_reg.fit(X_train_std,y_train_std)
```

## 学習モデルを保存

```python
with open('svm.pickle','wb') as f:
    pickle.dump(svm_reg,f)
```

## 保存した学習モデルを読み込み

```python
with open('svm.pickle','rb') as f:
    load_model=pickle.load(f)
```

```python
load_model.predict(X_test)
```
