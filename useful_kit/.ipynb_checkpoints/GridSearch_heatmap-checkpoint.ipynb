# GridSearchの結果をヒートマップで表示する
グリッドサーチでまずは、おおまかに値を絞り込み、徐々に狭い範囲でパラメータチューニングする際に役立つ。  
適切なパラメータの範囲にしないと意味がないので、気を付ける

```python
import sklearn
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from IPython.display import display
import seaborn as sns
```

## データセットの準備
## irisデータセット

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris=datasets.load_iris()
X=iris.data
y=iris.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

## GridSearchするモデルを作成

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svm=SVC(kernel='rbf')
grid_dict=[{'C':[10**i for i in range(-3,3)],'gamma':[10**i for i in range(-3,3)]}]

svm_clf=GridSearchCV(estimator=svm,param_grid=grid_dict,scoring='accuracy',cv=5)
svm_clf.fit(X_train,y_train)
```

## ヒートマップ表示する関数

```python
def plot_heatmap_from_grid(grid):
    # チューニング対象のパラメータを特定する。
    params = [k for k in grid.cv_results_.keys() if k.startswith('param_')]
    if len(params) != 2: raise Exception('grid has to have exact 2 parameters.') 

    # ヒートマップの行、列、値に使うキーを定義する。
    index = params[0]
    columns = params[1]
    values = 'mean_test_score'

    # gridから必要なキーのみを抽出する。
    df_dict = {k: grid.cv_results_[k] for k in grid.cv_results_.keys() & {index, columns, values}}

    # dictをDataFrameに変換してseabornでヒートマップを表示する。
    import pandas as pd
    df = pd.DataFrame(df_dict)
    data = df.pivot(index=index, columns=columns, values=values)
    import seaborn as sns
    sns.heatmap(data, annot=True,fmt='.3f')
```

## スコアが高い部分と低い部分が明確に分かる適切なヒートマップ

```python
plot_heatmap_from_grid(svm_clf)
```

## より狭い範囲で絞り込む

```python
svm=SVC(kernel='rbf')
grid_dict=[{'C':[10**i for i in range(-1,3)],'gamma':[0.01- 0.0001*i for i in range(-3,3)]}]

svm_clf=GridSearchCV(estimator=svm,param_grid=grid_dict,scoring='accuracy',cv=5)
svm_clf.fit(X_train,y_train)
plot_heatmap_from_grid(svm_clf)
```

## ダメ例　高いスコアのパラメータの絞り込みができない状態

```python
svm=SVC(kernel='rbf')
grid_dict=[{'C':[10**i for i in range(-5,0)],'gamma':[10**i for i in range(-5,0)]}]

svm_clf=GridSearchCV(estimator=svm,param_grid=grid_dict,scoring='accuracy',cv=5)
svm_clf.fit(X_train,y_train)
plot_heatmap_from_grid(svm_clf)
```
